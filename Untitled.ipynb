```python
import pandas as pd
```

```python
#とりあえずCSVを読み込むよ！

df = pd.read_csv("today.csv", index_col=False)
df.head()
```

```python
#要約統計量を見ていくよ！
```

```python
df.describe()
```

```python
# 標準偏差見るとだいぶばらついているので、ヒストグラムで可視化するよ
# まずはいいねから
```

```python
fav = df.fav
%matplotlib inline
fav.plot.hist()
```

```python
#よくわからないから、favが多い順に並べる
df = df.sort_values(by='fav', ascending=False)
```

```python
# 上位100件を見てみるよ！
df.head(100)
```

```python
# のちのちCSVで分析するかもだから、保存しておくよ
df.to_csv("fav.csv")
```

```python
# いいねされてシェアされるコンテンツを散布図で可視化するよ
# リファレンス: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.plot.html
# 300 X 300

df.plot(kind='scatter', x='RT', y='fav', xlim=[0,300], ylim=[0,300], title="Favorites and Retweets:300x300")
```

```python
# 200 X 200
df.plot(kind='scatter', x='RT', y='fav', xlim=[0,200], ylim=[0,200], title="Favorites and Retweets:200x200")
```

```python
# 100 X 100

df.plot(kind='scatter', x='RT', y='fav', xlim=[0,100], ylim=[0,100], title="Favorites and Retweets:100x100")
```

```python
# いいね数とRT数が多いツイートの特徴を分析したいので、いったんいいね数もRT数も多いと判断できないうんこツイートをすべて削除する
# いいね数が10未満のものをうんこツイートとして定義する

df = df[df.fav >= 10]
len(df)
```

```python
# うんこツイートなしでもう一度グラフで可視化してみる

df.plot(kind='scatter', x='RT', y='fav', xlim=[0,50], ylim=[10,200], title="Favorites and Retweets without Shit Tweets")
```

```python
# くそツイなしのデータをCSVでエクスポートする
# pandasでやるのだるいので、リツイート比率(RT/いいね)をエクセルで出してから再度noshit.csvを呼び出す
df.to_csv("noshit.csv")
```

```python
# 追加したrate (RT / fav)が入っているか確認する
import pandas as pd
df = pd.read_csv("noshit.csv", index_col=0)
df= df.sort_values(by=["rate"], ascending=False)
```

```python
# RT率が高いものを選ぶ
df.head(100)
```
